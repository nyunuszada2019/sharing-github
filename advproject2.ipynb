```python
import matplotlib
from tqdm import tqdm
import pandas as pd
import datetime
import matplotlib.pyplot as plt
path1 = 'phase-angle-1s_2019-09-01_2020-09-30.csv'
```

```python

datatypes = {'time': 'str',
              'Winterthur': 'float32', 
              'Büdingen': 'float32',
              'Schondorf': 'float32',
              'Herzogenrath': 'float32',
              'Bremen': 'float32',
              'Dresden': 'float32',
              'Lleida': 'float32',
              'Sibiu': 'float32', 
              'Belfort Cedex': 'float32',
              'Wien (SBA)' : 'float32'}

cols = list(datatypes.keys())
```

```python
chunksize = 5_000_000 # 5 million rows
```

```python
%%time
df_list = [] # list to hold the batch dataframe

for df_chunk in tqdm(pd.read_csv(path1, usecols=cols, dtype=datatypes, chunksize=chunksize)):
     
    # Neat trick from https://www.kaggle.com/btyuhas/bayesian-optimization-with-xgboost
    # Using parse_dates would be much slower!
    df_chunk['time'] = df_chunk['time'].str.slice(0, 16)
    df_chunk['time'] = pd.to_datetime(df_chunk['time'], utc=True, format='%Y-%m-%d %H:%M:%S')
    
    # Can process each chunk of dataframe here
    
    # Alternatively, append the chunk to list and merge all
    df_list.append(df_chunk) 
```

```python
# Merge all dataframes into one dataframe
phasor_df = pd.concat(df_list)

# Delete the dataframe list to release memory
del df_list

# See what I have loaded
phasor_df.info()
```

```python
# Null values
print(phasor_df.isnull().sum())
```

```python
#imputation with mean
phase_angle_imputed=phasor_df.fillna(phasor_df.mean())
```

```python
phase_angle_imputed.plot(grid=True,subplots=True,figsize=(20,20))
plt.xlabel("Time")
plt.legend(loc="upper left")
plt.tight_layout()
```
